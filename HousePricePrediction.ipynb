```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
import sklearn.datasets
```

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
house_price_dataset
```

```python
house_dataframe = pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
```

```python
house_dataframe['MedHouseVal']= house_price_dataset.target
```

```python
house_dataframe
```

```python
correlation = house_dataframe.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
x = house_dataframe.drop(['MedHouseVal'],axis=1)
y = house_dataframe['MedHouseVal']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
model = XGBRegressor()
model.fit(x_train,y_train)
```

```python
training_data_prediction = model.predict(x_train)
print(training_data_prediction)
```

```python
score_1 = metrics.r2_score(y_train,training_data_prediction)
print("R squared error",score_1)
score_2 = metrics.mean_absolute_error(y_train,training_data_prediction)
print("Mean absolute error",score_2)
```

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

```python
test_data_prediction = model.predict(x_test)
print(test_data_prediction)
```

```python
score_1 = metrics.r2_score(y_test,test_data_prediction)
print("R squared error",score_1)
score_2 = metrics.mean_absolute_error(y_test,test_data_prediction)
print("Mean absolute error",score_2)
```

```python
plt.scatter(y_test,test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
